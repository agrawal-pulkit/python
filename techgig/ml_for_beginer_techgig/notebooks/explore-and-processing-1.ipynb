```python
def credit_tag(range):
    if range == '95 TO 100': 
        return 'A'
    elif range == '90 TO 94': 
        return 'B'
    elif range == '85 TO 89': 
        return 'C'
    elif range == '80 TO 84': 
        return 'D'
    elif range == '75 TO 79': 
        return 'E'
    elif range == '70 TO 74': 
        return 'F'
    elif range == 'LESS THAN 70': 
        return 'F'
    elif range == 'PROFESSIONAL INDIVIDUAL': 
        return 'G'
    elif range == 'INSTITUTION': 
        return 'H'

df['CSR'] = df['credit_score_range'].map(lambda x: credit_tag(x))
```

```python
#create one-hot encoding 
df = pd.get_dummies(df, columns=['time_zone', 'credit_score', 'state', 'location_employee_code', 'employee_size_bin', 'CSR', 'store_location'])
```

```python
df.info()
```

```python
#drop unused columns
df.drop([ 'city', 'credit_score_range'], axis=1, inplace=True)
```

```python
#reorder columns
columns = [column for column in df.columns if column != 'total_sales']
columns = ['total_sales'] + columns
df = df[columns]
```

```python
#path for write processed data  
processed_data_path = os.path.join(os.path.pardir, 'data', 'processed')
write_train_data = os.path.join(processed_data_path, 'train_processed.csv')
write_test_data = os.path.join(processed_data_path, 'test_processed.csv')
```

```python
#write train data
df.loc[df.total_sales != -888].to_csv(write_train_data)

#write test data 
columns = [column for column in df.columns if column != 'total_sales']
df.loc[df.total_sales == -888, columns].to_csv(write_test_data)
```

```python
def get_submission_file(model, filename):
    #test data 
    test_X = df_test.as_matrix().astype('float')
    #prediction on test data 
    predictions = model.predict(test_X)
    #predicted data frames
    df_submission = pd.DataFrame({'PassengerId': df_test.index, 'total_sales': predictions})
    #set external file path
    submission_data_path = os.path.join(os.path.pardir, 'data', 'external')
    submission_file_path = os.path.join(submission_data_path, filename)
    #create csv file for baseline model prediction
    df_submission.to_csv(submission_file_path, index=False)
```
